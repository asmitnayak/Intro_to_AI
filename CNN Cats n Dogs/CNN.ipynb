```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten
from tensorflow.keras.layers import Conv2D, MaxPooling2D
import numpy as np
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
imgdatagen = ImageDataGenerator(rescale = 1.0/255)
training_set = imgdatagen.flow_from_directory('dog-cat-full-dataset-master/data/train', target_size = (64, 64), batch_size = 32, class_mode = 'binary')
test_set = imgdatagen.flow_from_directory('dog-cat-full-dataset-master/data/test', target_size = (64, 64), batch_size = 32, class_mode = 'binary')
```

```python
model = Sequential()
# Adding the initial convolution layer to my model
model.add(Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(64,64,3)))   # 64x64, RGB
model.add(MaxPooling2D(pool_size=(2, 2)))
# Adding three more
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.5))        # Reduces overfitting,
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.5))        # Reduces overfitting,
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())            # Converting into a 1D array
# Adding two hidden layer
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=128, activation='relu'))
# Adding the output layer
model.add(Dense(units=1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(training_set, steps_per_epoch = len(training_set), epochs = 25,
                    validation_data = test_set, validation_steps = len(test_set))
```

```python
import os
directory = 'cats_dogs/pred/'
corr = 0
for filename in os.listdir(directory):
    test_image = image.load_img(os.path.join(directory, filename), target_size = (64, 64))
    test_image = image.img_to_array(test_image)
    test_image = np.expand_dims(test_image, axis = 0)
    result = model.predict(test_image)
    prediction = ''
    if result[0][0] == 1:
        prediction = 'dog'
    else:
        prediction = 'cat'
    if prediction in filename:
        corr += 1
    # print(prediction, "\t"+filename)
print("Accuracy: ", corr*100/56.0)
```
